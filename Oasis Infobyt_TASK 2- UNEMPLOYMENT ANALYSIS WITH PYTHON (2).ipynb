#### Inferences
    1. As the latitude and longitude columns have no impact on Unemployment. So these two columns are not required.
    2. Using Region.1 column, mapped a New Region column. Hence, dropping the Region.1 column.
    3. In case of Area Column we observe missing values. Since we can not identified whether it is a Rural or Urban data, this column cannot be mapped. Hence, Area column will be dropped.

```python
#dropping unnecessary column
final_data = final_data.drop(['Region.1','longitude','latitude', 'Area'], axis=1)
```

```python
#rename column
final_data = final_data.rename(columns={'Region':'State',
                                       'New Region':'Region'})
```

```python
#handle the null value in Region Column 
final_data.Region = final_data.Region.fillna("North")
```
